1) Find celltype specific threshold for nCounts and percent.mt

```r
library(Seurat)
library(dplyr)

# Assuming 'data_rds' is your Seurat object
meta_data <- data_rds@meta.data

# Function to calculate mean, sd, and Z-score thresholds for each attribute with clear column names
calculate_stats <- function(data, attribute) {
  # Ensure the attribute column is numeric
  if (!is.numeric(data[[attribute]])) {
    stop(paste(attribute, "is not numeric. Please check the column data type."))
  }

  grouped_data <- data %>%
    group_by(celltype) %>%
    summarise(
      means = mean(.data[[attribute]], na.rm = TRUE),
      sds = sd(.data[[attribute]], na.rm = TRUE),
      .groups = "drop"
    )
  
  # Calculate Z-score thresholds
  grouped_data <- grouped_data %>%
    mutate(
      upper_threshold = means + 2 * sds,
      lower_threshold = means - 2 * sds
    )
  
  # Rename columns to include attribute name for clarity
  colnames(grouped_data)[-1] <- paste(colnames(grouped_data)[-1], attribute, sep = "_")
  
  return(grouped_data)
}

# Applying the function to each attribute
results_nCount_RNA <- calculate_stats(meta_data, "nCount_RNA")
results_nFeature_RNA <- calculate_stats(meta_data, "nFeature_RNA")
results_percent_mt <- calculate_stats(meta_data, "percent.mt")

# Merging results for easy viewing
results <- results_nCount_RNA %>%
  left_join(results_nFeature_RNA, by = "celltype") %>%
  left_join(results_percent_mt, by = "celltype")

# Print the results
print(results)
```

```r
library(Seurat)

# Define thresholds for each cell type (example values)
thresholds <- data.frame(
  celltype = c("Astrocyte", "DA", "Microglia"),
  nFeature_RNA_low = c(94, 0, 426),
  nFeature_RNA_high = c(5700, 4919, 4426),
  percent_mt_high = c(5, 7, 5)
)


# Initialize a vector to hold all the cells that meet the criteria across cell types
all_cells_to_keep <- character(0)

# Loop through each cell type to get the cells that meet the criteria
for (i in 1:nrow(thresholds)) {
  cell_type <- thresholds$celltype[i]
  feature_low <- thresholds$nFeature_RNA_low[i]
  feature_high <- thresholds$nFeature_RNA_high[i]
  mt_high <- thresholds$percent_mt_high[i]

  # Find cells that meet the criteria
  cells_this_type <- WhichCells(
    object = data_rds,
      expression = nFeature_RNA > feature_low &
                                                     nFeature_RNA < feature_high &
                                                     percent.mt < mt_high & 
                                                     celltype == cell_type)

  # Append to the list of all cells to keep
  all_cells_to_keep <- c(all_cells_to_keep, cells_this_type)
}

# Filter the original Seurat object to keep only the desired cells
# data_rds <- subset(data_rds, cells = all_cells_to_keep)

# Check the updated Seurat object
# print(data_rds)
```
